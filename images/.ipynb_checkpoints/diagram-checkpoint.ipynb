```python
from matplotlib import pyplot as plt
```

```python
before_data = []
with open("native_stats", "r") as file:
    for line in file:
        before_data.append(float(line))
```

```python
after_data = []
with open("cpu_stats", "r") as file:
    for line in file:
        after_data.append(float(line))
```

```python
before_mean = sum(before_data) / len(before_data)
after_mean = sum(after_data) / len(after_data)
coeff = after_mean / before_mean
```

```python
plt.figure(figsize=(20,10))
plt.bar("Native", before_mean, label="Native x1.0")
plt.bar("CPU Emulator", after_mean, label="CPU Emulator x" + str(round(coeff, 2)))
plt.ylabel("Time, ms", fontsize=25);
plt.legend(prop={'size': 25})
plt.title("Time, ms per 1000 countings of factorial of 31", fontsize=30)
plt.savefig("diagram.png")
```
